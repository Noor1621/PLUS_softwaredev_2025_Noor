```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create figure and axis with geographic projection
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Zoom to Austria area: [west_lon, east_lon, south_lat, north_lat]
ax.set_extent([9, 17, 46, 49], crs=ccrs.PlateCarree())

# Add features for better visualization
ax.coastlines(resolution='10m')  # detailed coastlines
ax.add_feature(cfeature.BORDERS, linestyle='--')
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS)

# Add gridlines with labels
gl = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
gl.top_labels = False
gl.right_labels = False

# Plot Salzburg point
salzburg_lon, salzburg_lat = 13.0550, 47.8095
ax.plot(salzburg_lon, salzburg_lat, 'ro', markersize=10, transform=ccrs.PlateCarree(), label='Salzburg')

# Add legend and title
ax.legend(loc='upper left')
plt.title("Zoomed Map Around Salzburg, Austria")

plt.show()
```

```python
# Simple Geo Visualization with Matplotlib and Cartopy

In this notebook, we use Matplotlib with Cartopy to create a geographic map focused on Austria.

We added detailed coastlines, country borders, lakes, rivers, and gridlines for better visualization.

We zoomed into Austria using the `set_extent` function and marked the location of Salzburg with a red dot.

Cartopy documentation: https://scitools.org.uk/cartopy/docs/latest/
```
